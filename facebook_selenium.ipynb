```python
# https://jakevdp.github.io/blog/2017/12/05/installing-python-packages-from-jupyter/
# https://gist.github.com/u0401006/5914d98393484c40a74e1db3154df34c
# import sys
# !{sys.executable} -m pip install selenium
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import time as tm
import pandas as pd
from random2 import randrange
```

```python
#設定webdriver的參數，這個prefs是關掉Facebook通知

chrome_options = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2}

chrome_options.add_experimental_option("prefs",prefs)
```

```python
#https://medium.com/@Epicure1709/%E4%BD%BF%E7%94%A8python%E7%9A%84selenium%E6%99%82%E9%81%87%E5%88%B0%E7%9A%84%E4%B8%80%E4%BA%9B%E5%B0%8F%E5%95%8F%E9%A1%8C-7fb5de198ff7
#指定路徑
driver=webdriver.Chrome('/Users/dtseng02/Documents/CRM/cralwer/selenium/chromedriver', options=chrome_options)
# '/Users/dtseng02/Documents/CRM/cralwer/selenium/chromedriver'
```

```python
#假瀏覽器打開的網址
url_raw = "https://m.facebook.com/ntusociology/"
driver.get(url_raw)
```

```python
#開4個list暫存解析的資料
p_links, p_time, p_content, p_interaction = [], [], [], []
```

```python
counter = 0
while counter < 20:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    tm.sleep(randrange(0,3))
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    tm.sleep(randrange(0,3))

    soup = BeautifulSoup(driver.page_source,"lxml")
    
    post_div = soup.find_all("article")

	#抓到每一則貼文的層級    
    for j,i in enumerate(post_div):

    	#貼文內容
        try:
            p_content.append(''.join([k.text for k in i.find('header').find_next_siblings()]))
        except:
            p_content.append('')
            print('content'+str(j))

        #貼文時間
        try:
            p_time.append(i.find("header").find("abbr").text)
        except AttributeError:
            try:
                p_time.append(i.find("header").find_next_sibling().find('abbr').text)
            except:
                p_time.append('')
                print('time'+str(j))

        #貼文連結
        try:
            p_links.append(i.find("header").find('h3').find_next_sibling().find('a')['href'])
        except:
            try:
                p_links.append(i.find("header").find('a')['href'])
            except:
                p_links.append('')
                print('link'+str(j))

        #貼文互動數據
        try:
            p_interaction.append(i.find('footer').find("span").text)
        except:
            p_interaction.append('')
            print('reaction'+str(j))
    
    counter += 1
    if counter % 500 == 1:
        df = pd.DataFrame()
        columns = ['p_content','p_time','p_links','p_interaction']
        cols = [p_content, p_time, p_links, p_interaction]
        for i,j in zip(columns,cols):
            df[i] = j
        df.to_csv('tsai-fb-re-m.csv',sep='\t',encoding='utf-8')
```
